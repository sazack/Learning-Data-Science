```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_GDP"
response = requests.get(url)
pagetext = response.text
```

```python
soup = BeautifulSoup(pagetext,'lxml')
# print(soup.prettify())
```

```python
table = soup.find('table',{'class':"wikitable sortable"})
```

```python
df = pd.read_html(str(table))[0]
```

```python
df.head()
```

```python
new_header = df.iloc[0]
df = df[2:]
df.columns = new_header
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
RegionCount  = df['Region'].value_counts()
RegionCount.plot(kind='bar', figsize =(10,8), title="Number of State based on region")
```

```python
NEamount = df[df['Region']== 'Northeast']['2017'].astype(int).values.sum()
Wamount = df[df['Region']== 'West']['2017'].astype(int).values.sum()
MWamount = df[df['Region']== 'Midwest']['2017'].astype(int).values.sum()
Samount = df[df['Region']== 'South']['2017'].astype(int).values.sum()

RegAmtDF = pd.DataFrame([Samount,Wamount,MWamount,NEamount])
RegAmtDF.index = ['South', 'West', 'Mid West','North East']
RegAmtDF.columns = ['Amount']
RegAmtDF.head()
```

```python
RegAmtDF.plot(kind='pie', title="Amount of GDP based on region", figsize=(15,8), subplots=True)
```
